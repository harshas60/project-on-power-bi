# E-Commerce Sales Analytics Notebook

This notebook demonstrates interactive analysis of the e-commerce sales dataset, including revenue trend analysis, customer segmentation, and product performance metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
data_path = "data/transactions_clean.csv"
df = pd.read_csv(data_path, parse_dates=["transaction_date"])
df["year_month"] = df["transaction_date"].dt.to_period("M").astype(str)

df.head()
```

## Revenue Trend and Regional Performance

Plot the monthly revenue trend and compare revenue across regions.

```python
revenue_by_month = df.groupby("year_month")["revenue"].sum().reset_index()
revenue_by_region = df.groupby("region")["revenue"].sum().reset_index().sort_values("revenue", ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
sns.lineplot(data=revenue_by_month, x="year_month", y="revenue", marker="o", ax=axes[0])
axes[0].set_title("Monthly Revenue Trend")
axes[0].tick_params(axis="x", rotation=45)
sns.barplot(data=revenue_by_region, x="revenue", y="region", palette="viridis", ax=axes[1])
axes[1].set_title("Revenue by Region")
plt.tight_layout()
plt.show()
```

## Customer Segmentation and Revenue Concentration

Calculate customer lifetime revenue, segment customers, and estimate the 23% revenue concentration insight.
